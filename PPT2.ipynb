```python
from IPython.display import HTML, display
from pathlib import Path
import re

html_dir = Path("htmls")

def natural_key(f):
    """Devuelve un número para ordenar archivos tipo 1.html, 2.html ..."""
    m = re.search(r'(\d+)', f.name)
    return int(m.group(1)) if m else float('inf')

# Recorre todos los HTML en orden natural
for f in sorted(html_dir.glob("*.html"), key=natural_key):
    html_content = f.read_text(encoding="utf-8")
    
    # Extraemos solo el <body> para evitar problemas de <html> y <head>
    body_start = html_content.find("<body>")
    body_end = html_content.find("</body>")
    if body_start != -1 and body_end != -1:
        body_content = html_content[body_start + 6:body_end]
    else:
        body_content = html_content  # fallback, si no tiene <body>

    display(HTML(f"<div class='slide'>{body_content}</div>"))
```
