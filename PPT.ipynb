```python
from IPython.display import IFrame, display
from pathlib import Path
import re

html_dir = Path("htmls")

def natural_key(f):
    """
    Devuelve un número para ordenar archivos tipo 1.html, 2.html, ..., 17.html
    Si no encuentra número, lo manda al final.
    """
    m = re.search(r'(\d+)', f.name)
    return int(m.group(1)) if m else float('inf')

# Recorre todos los HTML en orden natural
for f in sorted(html_dir.glob("*.html"), key=natural_key):
    display(
        IFrame(
            src=str(f),
            width=1400,
            height=800
        )
    )
```
